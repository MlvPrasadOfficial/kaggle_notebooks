```python
# Example: Creating a DataFrame from a dictionary
import pandas as pd

# Create a DataFrame from a dictionary with default index and columns
data = {'Name': ['John', 'Emily', 'Michael'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data)
df

```
